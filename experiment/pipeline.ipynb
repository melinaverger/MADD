```python
"""
Created on: Wed. 5 Oct. 2022
Updated on: Thu. 13 Oct. 2022
Updated on: Fri. 14 Oct. 2022
Updated on: Mon. 31 Oct. 2022
Updated on: Mon. 14 Nov. 2022
Updated on: Wed. 4 Jan. 2023
Updated on: Thu. 5 Jan. 2023
Updated on: Fri. 6 Jan. 2023
Updated on: Wed. 11 Jan. 2023
Author: MÃ©lina Verger

Run all `pipeline.ipynb`. You will have to enter:
* the data set you chose,
* if you want to reduce your data set to course 'BBB' or course 'FFF' data,
* the train-test split ratio you want between 70-30 or 80-20,
* the sensitive attributes you should have identified,
* the classifier models to study.
"""

# To exit script
from sys import exit

# For data manipulation
import pandas as pd
pd.set_option('display.max_columns', None)

# To make the train-test split
from sklearn.model_selection import train_test_split

# To save the variables
import pickle

# To interact with the operating system
import os

# ML models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import ComplementNB, MultinomialNB
```

## Load a numerical and scaled data set

```python
# Load the chosen data set (num and scaled)

user_response1 = input("Do you want to work with stInfo, stAll or stClick set (scaled)? (Write either stInfo, stAll or stClick.)")
if user_response1 == "stInfo":
    data_num_scaled = pd.read_csv("../data/studentInfo_num_scaled.csv")
elif user_response1 == "stAll":
    data_num_scaled = pd.read_csv("../data/studentAll_num_scaled.csv")
elif user_response1 == "stClick":
    data_num_scaled = pd.read_csv("../data/studentClick_num_scaled.csv")
else:
    print("Invalid choice.")
    exit()

print(user_response1)
```

```python
data_num_scaled
```

```python
# Either binarize age_band

def binarize_age(x):
    if x == 0:
        return 0
    elif x == 0.5:
        return 1
    elif x == 1:
        return 1

user_response_age = input("Do you want to binarize the feature 'age_band' (it will be renamed as 'age')? (y/n)")
if user_response_age == "y":
    data_num_scaled.age_band = data_num_scaled.age_band.apply(binarize_age)
```

```python
DATA = user_response1
pickle.dump(DATA, open("../data/DATA", "wb"))
```

## Reduction data set to courses

```python
# Choose a course

data_num_scaled_reduced = data_num_scaled

user_response3 = input("Do you want to reduce the data set to the course 'BBB'? (y/n)")
if user_response3 == 'y':
    data_num_scaled_reduced = data_num_scaled_reduced.loc[data_num_scaled_reduced["code_module_BBB"] == 1]
    print("Data reduction to course 'BBB'.", flush=True)
    print(f"Data shape: {data_num_scaled_reduced.shape}\n", flush=True)
elif user_response3 == 'n':
    print("No data set reduction to course 'BBB'.", flush=True)
else:
    print("Invalid choice.")
    exit()

user_response3bis = input("Do you want to reduce the data set to the course 'FFF'? (y/n)")
if user_response3bis == 'y':
    data_num_scaled_reduced = data_num_scaled_reduced.loc[data_num_scaled_reduced["code_module_FFF"] == 1]
    print("Data reduction to course 'FFF'.", flush=True)
    print(f"Data shape: {data_num_scaled_reduced.shape}\n", flush=True)
elif user_response3bis == 'n':
    print("No data set reduction to course 'FFF'.", flush=True)
else:
    print("Invalid choice.")
    exit()
```

```python
data_num_scaled_reduced.shape
```

```python
# Keep the interesting features

if user_response1 == "stInfo":
    data_num_scaled_reduced = data_num_scaled_reduced[["gender", "highest_education", "imd_band", "age_band",  "disability", "num_of_prev_attempts", "studied_credits", "final_result"]]

if user_response1 == "stAll":
    data_num_scaled_reduced = data_num_scaled_reduced[["gender", "highest_education", "imd_band", "age_band",  "disability", "num_of_prev_attempts", "studied_credits", "sum_click", "weighted_score", "final_result"]]

if user_response1 == "stClick":
    data_num_scaled_reduced = data_num_scaled_reduced[["gender", "highest_education", "imd_band", "age_band",  "disability", "num_of_prev_attempts", "studied_credits", "sum_click", "final_result"]]
```

```python
data_num_scaled_reduced.shape
```

```python
# Rename "age_band" and "imd_band"

data_num_scaled_reduced.rename(columns={"age_band": "age"}, inplace=True)
data_num_scaled_reduced.rename(columns={"imd_band": "poverty"}, inplace=True)
```

## Machine learning (preparation)

### X-y split

```python
X = data_num_scaled_reduced.drop(columns="final_result")
y = data_num_scaled_reduced["final_result"]
```

### Train-test split

```python
# Choose a train-test split

user_response2 = input("Do you want 70-30 or 80-20 train-test split? (Write either 7030 or 8020.)")
if user_response2 == "7030" or user_response2 == "8020":
    SPLIT = user_response2
else:
    print("Invalid choice.")
    exit()

print(user_response2)
```

```python
pickle.dump(SPLIT, open("../data/SPLIT", "wb"))
```

```python
if SPLIT == "7030":
    ratio = 0.30
elif SPLIT == "8020":
    ratio = 0.20
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=ratio, random_state=0, stratify=y, shuffle=True)
```

```python
# Gender imbalance

round(X_train["gender"].value_counts()[0] / (X_train["gender"].value_counts()[0] + X_train["gender"].value_counts()[1]) * 100, 2)
```

```python
# Positive class imbalance

round(y_train.value_counts()[1] / (y_train.value_counts()[1] + y_train.value_counts()[0]) * 100, 2)
```

```python
pd.DataFrame(X_train).to_csv("../data/X_train" + "_" + DATA + "_" + SPLIT + ".csv", index=False)
pd.DataFrame(X_test).to_csv("../data/X_test" + "_" + DATA + "_" + SPLIT + ".csv", index=False)
pd.DataFrame(y_train).to_csv("../data/y_train" + "_" + DATA + "_" + SPLIT + ".csv", index=False)
pd.DataFrame(y_test).to_csv("../data/y_test" + "_" + DATA + "_" + SPLIT + ".csv", index=False)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Choose the sensitive features

```python
user_response4 = int(input(f"How many sensitive features do you consider into {list(X_train.columns)}? (Write a number greater than 0.)"))

L_sensitive_features = list()

if user_response4 == 1:
    feature_name = input(f"Write a feature name belonging to {list(X_train.columns)}: ")
    if feature_name in X_train.columns:
        L_sensitive_features.append(feature_name)
    else:
        print("Invalid name.")
        exit()

else:
    for i in range(user_response4):
        feature_name = input(f"Write the feature name #{i + 1} belonging to {list(X_train.columns)}: ")
        if feature_name in X_train.columns:
            L_sensitive_features.append(feature_name)
        else:
            print("Invalid name.")
            exit()

print("Sensitive features: ", L_sensitive_features)
```

```python
SFEATURES = L_sensitive_features
pickle.dump(SFEATURES, open("../data/SFEATURES", "wb"))
```

## Select the classifier models

```python
model_names = []
models = dict()

user_response_lr = input("Do you want to use a logistic regression classifier (LR)? (y/n)")
if user_response_lr == "y":
    clf_lr = LogisticRegression(random_state=0)
    model_names.append("a logistic regression")
    models["clf_lr"] = clf_lr

user_response_kn = input("Do you want to use a k-nearest neighbors classifier (KN)? (y/n)")
if user_response_kn == "y":
    clf_kn = KNeighborsClassifier()
    model_names.append("a k-nearest neighbors")
    models["clf_kn"] = clf_kn

user_response_dt = input("Do you want to use a decision tree classifier (DT)? (y/n)")
if user_response_dt == "y":
    clf_dt = DecisionTreeClassifier(random_state=0)
    model_names.append("a decision tree")
    models["clf_dt"] = clf_dt

user_response_rf = input("Do you want to use a random forest classifier (RF)? (y/n)")
if user_response_rf == "y":
    clf_rf = RandomForestClassifier(random_state=0)
    model_names.append("a random forest")
    models["clf_rf"] = clf_rf

user_response_svc = input("Do you want to use a linear support vector classifier (SVC)? (y/n)")
if user_response_svc == "y":
    clf_svc = LinearSVC(random_state=0)
    model_names.append("a linear support vector")
    models["clf_svc"] = clf_svc

user_response_cnb = input("Do you want to use a complement naive bayes classifier (CNB)? (y/n)")
if user_response_cnb == "y":
    clf_cnb = ComplementNB()
    model_names.append("a complement naive bayes")
    models["clf_cnb"] = clf_cnb

user_response_mnb = input("Do you want to use a multinomial naive bayes classifier (MNB)? (y/n)")
if user_response_mnb == "y":
    clf_mnb = MultinomialNB()
    model_names.append("a multinomial naive bayes")
    models["clf_mnb"] = clf_mnb

if len(model_names) == 0:
    print("You chose none of these models. Please implement one.")
else:
    print(f"You chose {', '.join(model_names)} classifiers.", flush=True)
```

```python
# Train and save trained models

for mod_name in models:
    mod = models[mod_name]  # the model object
    models[mod_name] = mod.fit(X_train, y_train)  # update with trained models
    

MODELS = models
pickle.dump(MODELS, open("../data/MODELS", "wb"))
```
