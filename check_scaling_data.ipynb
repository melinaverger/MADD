```python
"""
Created on: Tue. 9 Aug. 2022
Author: Mélina Verger
"""

import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
studentInfo_num = pd.read_csv("./data/studentInfo_num.csv")
studentInfo_num_scaled = pd.read_csv("./data/studentInfo_num_scaled.csv")
```

```python
studentInfo_num
```

```python
studentInfo_num_scaled
```

```python
for column in studentInfo_num.columns:
    if column == "highest_education" or column == "age_band" or column == "num_of_prev_attempts" or column == "studied_credits":
        # plot both together to compare
        fig, ax=plt.subplots(1,2, figsize=(16, 6))
        sns.histplot(studentInfo_num[column], ax=ax[0], color="blue")
        ax[0].set_title("Original Data")
        sns.histplot(studentInfo_num_scaled[column], ax=ax[1], color="blue")
        ax[1].set_title("Scaled data")
        fig.suptitle("Feature: " + column, fontsize=16)
        plt.show()
```
