```python
# Load the chosen data set

user_response = input("Do you want to check stInfo or stAll set after preprocessing? (Write either stInfo or stAll.)")
if user_response == "stInfo":
    data_num = pd.read_csv("./data/studentInfo_num.csv")
    data_num_scaled = pd.read_csv("./data/studentInfo_num_scaled.csv")
elif user_response == "stAll":
    data_num = pd.read_csv("./data/studentAll_num.csv")
    data_num_scaled = pd.read_csv("./data/studentAll_num_scaled.csv")
else:
    print("Invalid choice.")
    exit()
```

```python
for column in data_num.columns:
    if column == "highest_education" or column == "age_band" or column == "num_of_prev_attempts" or column == "studied_credits" or column == "sum_click" or column == "weighted_score":
        # plot both together to compare
        fig, ax=plt.subplots(1,2, figsize=(16, 6))
        sns.histplot(data_num[column], ax=ax[0], color="blue")
        ax[0].set_title("Original Data")
        sns.histplot(data_num_scaled[column], ax=ax[1], color="blue")
        ax[1].set_title("Scaled data")
        fig.suptitle("Feature: " + column, fontsize=22)
        plt.show()
```
